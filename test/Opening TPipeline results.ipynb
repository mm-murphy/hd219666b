```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.load('FitTransitSingle_Spectra.npz', allow_pickle=True)
results = dict(results)
values = results['FitOutput'].item()
```

```python
depths = np.array([])
depth_errs = np.array([])
for key in values.keys():
    if 'depth' in key:
        depths = np.append(depths, values[key][0])
        err = np.mean([values[key][1], values[key][2]])
        depth_errs = np.append(depth_errs, err)
```

```python
waves = np.linspace(1.1, 1.6, len(depths))
plt.errorbar(waves, depths*1.e6, yerr=depth_errs*1.e6)
plt.show()
```

```python
tp_td_depths = np.copy(depths)
tp_td_deptherrs = np.copy(depth_errs)

%store tp_td_depths
%store tp_td_deptherrs
```
