Taking a look at the response function for G141 downloaded from http://svo2.cab.inta-csic.es/svo/theory/fps3/index.php?id=HST/WFC3_IR.G141

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = '/home/matthew/data/calibrations/HST_WFC3_IR_G141_response.dat'
dat = np.loadtxt(file, unpack=True)
wavelength = dat[0]   # wavelength in [AA]
response = dat[1]     # response function value
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(wavelength, response)
ax.set_xlabel('Wavelength [Angstrom]', fontsize=12)
ax.set_ylabel('Filter Response', fontsize=12)
plt.show()
```
