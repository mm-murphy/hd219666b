In this notebook, I extract the broadband flux time series as well as the spectrophometric time series (i.e. the raw transmission spectrum) from our HST data. 

```python
# Importing modules we need:
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
import astropy
import astropy.convolution as ac
from astropy.io import fits
from scipy.interpolate import interp1d
# Importing HST extraction functions
#import HST_Extract_Funcs as hst
import HST_extraction as ext
import gif
```

Loading in the dictionary containing the file paths to our data

```python
FPload = np.load('FilePaths.npz', allow_pickle=True)
datapaths = FPload['filedict'].item()   # dictionary containing paths to the image files
```

```python
# Loading in the first two exposures in orbit 2
image0path = datapaths['visit3']['orbit2']['grism'][0]
image1path = datapaths['visit3']['orbit2']['grism'][1]
# and this orbit's direct image
directimagepath = datapaths['visit3']['orbit2']['direct']
# computing the wavelength solution (needed for sub-exposure separation function)
wavesolution = ext.WavelengthSolution(directimagepath)
# Extracting their sub-exposures
image0_subexpdetails = ext.CreateSubExps(image0path, wavesolution)
image0_subexps = image0_subexpdetails[0]
image0_scandir = image0_subexpdetails[4]

image1_subexpdetails = ext.CreateSubExps(image1path, wavesolution)
image1_subexps = image1_subexpdetails[0]
image1_scandir = image1_subexpdetails[4]
```

```python
image0_subexps.shape
```

```python
print(image0_scandir)
for subexp in range(image0_subexps.shape[0]):
    subexp_img = image0_subexps[subexp]
    plt.imshow(subexp_img, origin='lower', cmap='magma')
    plt.show()
```

```python
subarray_size = image0_subexps.shape[1]
image0frames = np.zeros(image0_subexps.shape)
for subexp in range(image0_subexps.shape[0]):
    j = 0
    while j <= subexp:
        image0frames[subexp] += image0_subexps[j]
        j += 1
    
@gif.frame
def plotter(frames, frame):
    frame_img = frames[frame]
    plt.imshow(frame_img, origin='lower', cmap='magma')
    plt.title('Scan Direction %d'%(image0_scandir))
    plt.xlabel('Dispersion Axis')
    plt.ylabel('Scan Axis')
    
gif_frames = []
for frame in range(image0frames.shape[0]):
    gif_frames.append(plotter(image0frames, frame))
    
gif.save(gif_frames, 'orbit2image0_scandir1.gif', duration=5, unit='s', between='startend')
```

```python
subarray_size = image1_subexps.shape[1]
image1frames = np.zeros(image1_subexps.shape)
for subexp in range(image1_subexps.shape[0]):
    j = 0
    while j <= subexp:
        image1frames[subexp] += image1_subexps[j]
        j += 1
    
@gif.frame
def plotter(frames, frame):
    frame_img = frames[frame]
    plt.imshow(frame_img, origin='lower', cmap='magma')
    plt.title('Scan Direction %d'%(image1_scandir))
    plt.xlabel('Dispersion Axis')
    plt.ylabel('Scan Axis')
    
gif_frames = []
for frame in range(image1frames.shape[0]):
    gif_frames.append(plotter(image1frames, frame))
    
gif.save(gif_frames, 'orbit2image1_scandir0.gif', duration=5, unit='s', between='startend')
```
