```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Thomas' orbit 2 data ... for comparison
# Full first image (including all subexposures) of orbit 2 
Tfile = np.load('thomas_images/idz206jwq_ima.fits.npz')  # first image of orbit 2
raw = Tfile['prebkgd']                    # raw image
bksub = Tfile['bkgdsubbed']               # post- background subtraction 
flatted = Tfile['calibrated']             # post- flatfielding
```

```python
for img in range(flatted.shape[0]):
    plt.imshow(flatted[img], origin='lower', vmin=0)
    plt.title('Subexpsoure %d'%(img))
    plt.show()
```
