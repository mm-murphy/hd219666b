```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# properties of HD 219666 b
HDmass, HDmasserr = 18.0, 1.2  # [earth masses]
HDradius, HDradiuserr = 5.05, 0.05 # [earth radii], measured in HST/WFC3 G141 bandpass
```

```python
fig, ax = plt.subplots(figsize=(8,4), nrows=1)
ax.errorbar(HDmass, HDradius, xerr=HDmasserr, yerr=HDradiuserr,
           ls=None, marker='o', ms=5, c='blue')
ax.text(HDmass+0.5, HDradius-0.2, 'HD 219666 b', c='blue', fontsize=8, fontweight='heavy')
ax.set_xlim(0., 25.)  # mass axis
ax.set_xlabel(r'Mass [M$_\bigoplus$]', fontsize=10)
ax.set_ylim(1., 6.) # radius axis
ax.set_ylabel(r'Radius [R$_\bigoplus$]', fontsize=10)
#plt.savefig('figures/MRdiagram.png', dpi=400, bbox_inches='tight')
plt.show()
```
